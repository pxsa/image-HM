```python
import cv2
from matplotlib import pyplot as plt
```

```python
pathes = ['assignment/Medium/panguan.jpg', 'assignment/Hard/288024.jpg', 'assignment/Hard/168084.jpg']
images = []

for path in pathes:
    images.append(cv2.imread(path))
```

```python
inter_nearst_zoomed_images = []
inter_leaner_zoomed_images = []

for image in images:
    inter_nearst_zoomed_images.append(cv2.resize(image, dsize=(0, 0), fx=2, fy=2, interpolation=cv2.INTER_NEAREST))
    inter_leaner_zoomed_images.append(cv2.resize(image, dsize=(0, 0), fx=2, fy=2, interpolation=cv2.INTER_LINEAR))
```

```python
figure, axis = plt.subplots(nrows=3, ncols=3, figsize=(12, 12))
for i in range(len(images)):
    axis[i, 0].imshow(images[i])
    axis[i, 0].set_title('Original')

    axis[i, 1].imshow(inter_nearst_zoomed_images[i])
    axis[i, 1].set_title('Nearst method')

    axis[i, 2].imshow(inter_leaner_zoomed_images[i])
    axis[i, 2].set_title('Linear method')

```
