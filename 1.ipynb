```python
import cv2
import matplotlib.pyplot as plt
```

```python
files = ['assignment/Hard/393035.jpg', 'assignment/Hard/288024.jpg', 'assignment/Hard/223004.jpg', 'assignment/Hard/87015.jpg', 'assignment/Easy/29030.jpg']
```

```python
images = []
for file in files:
    image = cv2.imread(file, 0)
    # image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    images.append(image)
```

```python
for img in images:
    print(img.shape)
```

```python

figure, axis = plt.subplots(nrows=1, ncols=len(images), figsize=(12,4))
for idx, image in enumerate(images):
    axis[idx].imshow(image, cmap='gray')
    axis[idx].set_title(f"{idx+1}")
    axis[idx].set_axis_off()

figure.tight_layout()
    
```

```python
# find frequency of pixels in range 0-255
hist_images = []
for image in images:
    histr = cv2.calcHist([image],[0],None,[256],[0,256])
    hist_images.append(histr)
```

```python
figure, axis = plt.subplots(nrows=1, ncols=len(images), figsize=(12, 4))
for idx, image in enumerate(hist_images):
    axis[idx].plot(image)
    axis[idx].set_title(f"{idx+1}")
    
figure.tight_layout()
```

```python
f, a = plt.subplots(nrows=1, ncols=len(images), figsize=(12, 4))

for idx,ax in enumerate(a):
    ax.hist(images[idx].ravel(), 256, [0, 256])
    ax.set_title(f'{idx+1}')
plt.tight_layout()
```

```python
after_equalization_images = []
for idx, image in enumerate(images):
    new_image = cv2.equalizeHist(image)
    after_equalization_images.append(new_image)
```

```python
f, a = plt.subplots(nrows=1, ncols=len(images), figsize=(12, 4))

for idx,ax in enumerate(a):
    ax.hist(after_equalization_images[idx].ravel(), 256, [0, 256])
    ax.set_title(f'{idx+1}')
plt.tight_layout()
```

```python
figure, axis = plt.subplots(nrows=1, ncols=len(images), figsize=(12, 4))
for idx, image in enumerate(after_equalization_images):
    # image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    axis[idx].plot(image)
    axis[idx].set_title(f"{idx+1}")
    
figure.tight_layout()
```

```python
figure, axis = plt.subplots(nrows=1, ncols=len(images), figsize=(12,4))
for idx, image in enumerate(after_equalization_images):
    axis[idx].imshow(image,  cmap='gray')
    axis[idx].set_title(f"{idx+1}")
    axis[idx].set_axis_off()

figure.tight_layout()
```

```python
all = images + after_equalization_images

figure, axis = plt.subplots(nrows=2, ncols=len(images), figsize=(12, 4))

for idx, image in enumerate(all):
    axis[idx // len(images), idx % len(images)].imshow(image, cmap='gray')
    axis[idx // len(images), idx % len(images)].set_title(f"{idx+1}")
    axis[idx // len(images), idx % len(images)].set_axis_off()

figure.tight_layout()
```
