```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# Read an image 
image_names = ['Hard/288024.jpg', 'Hard/393035.jpg', 'Hard/223004.jpg', 'Easy/29030.jpg']
images = []

for image_name in image_names:
    image = cv2.imread('assignment/' + image_name)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    images.append(image)
```

```python
for idx, image in enumerate(images):
    plt.subplot(1, 4, idx+1)
    plt.axis('off')
    plt.title(f'{idx+1}')
    plt.imshow(image)
```

```python
# Apply log transformation method 
log_images = []

for image in images:
    c = 255 / np.log(np.uint32(1) + np.max(image)) 
    log_image = c * (np.log(image + np.uint32(1), dtype=np.float32)) 
    # Specify the data type so that 
    # float value will be converted to int 
    log_image = np.array(log_image, dtype = np.uint32)
    log_images.append(log_image)

```

```python
all_images = images + log_images
```

```python
a = 3
for idx, image in enumerate(all_images):

    if idx < 4:
        pos = idx * 2  
    else:
        pos = idx - a
        a -= 1

    plt.subplot(2, 4, pos+1)
    plt.axis('off')
    # plt.title(f'{idx+1}')
    plt.imshow(image)
```
